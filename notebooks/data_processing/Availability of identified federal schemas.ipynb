From the set of unique, federally-hosted schemas, how many are currently accessible?

(noting, of course, that these are not always resolvable to start)

This is just a quick test - HEAD requests against our list of 200 schemas, assuming that those are the parent node and that we will not traverse any includes. And, for now, we won't be handling SSL issues or redirects.

```python
import os
import requests
```

```python
responses = {}
with open('outputs/federal_schemas_sorted.txt', 'r') as f:
    for line in f.readlines():
        url = line.strip()
        
        try:
            req = requests.head(url)
        except Exception as ex:
            responses[url] = {'error': str(ex), 'status': '-9999'}
            continue
        
        responses[url] = {'status': req.status_code}
        
```

```python
# save for reference later
import json
with open('outputs/federal_linkrot_20150925.json', 'w') as f:
    f.write(json.dumps(responses, indent=4))
```

```python
# let's do some quick aggregation
from collections import defaultdict

agg = defaultdict(int)
for response in responses.itervalues():
    key = response['status']
    agg[key] += 1
    
print agg
```

28 are gone or inaccessible.
Another 78 are redirected (this *does not* necessarily mean a redirect to a usable schema).
And the remaining 98 are 200.
